### Class imbalance

```python
#true positive, true negative, false positive, false negative

actual_positive = (y_val==1)
actual_negative = (y_val==0)
```

```python
t = 0.5
predict_positive = (y_pred >= t)
predict_negative = (y_pred < t)
```

```python
tp = (predict_positive & actual_positive).sum()
tn = (predict_negative & actual_negative).sum()
```

```python
fp = (predict_positive & actual_negative).sum()
fn = (predict_negative & actual_positive).sum()
```

```python
fp, fn
```

```python
#confustion matrix first row tn, fp; second row fn, tp

confusion_matrix = np.array([
    [tn, fp],
    [fn, tp]
])
```

```python
(confusion_matrix / confusion_matrix.sum()).round(2)
```

### Precision and recall

```python
# accurracy 
(tp + tn) / (tp + tn + fp + fn)
```

```python
# precision -> fraction of positive predictions that is correct
p = tp/(tp+fp)
p
```

```python
# recall -> fraction of correctly indentified positive examples
r = tp / (tp + fn)
r
```

### ROC (Receiver operating characteristics)

```python
#fale positive rate FPR and true positive rate TPR
#FPR = FP / (TN + FP) -> minimize
#TPR = TP / (TP + FN) -> maximize
```

```python
tpr = tp / (tp + fn)
fpr = fp / (fp + tn)
```

```python
scores = []
thresholds = np.linspace(0, 1, 101)

for t in thresholds:
    actual_positive = (y_val==1)
    actual_negative = (y_val==0)

    predict_positive = (y_pred >= t)
    predict_negative = (y_pred < t)
    
    tp = (predict_positive & actual_positive).sum()
    tn = (predict_negative & actual_negative).sum()
    fp = (predict_positive & actual_negative).sum()
    fn = (predict_negative & actual_positive).sum()

    scores.append((t, tp, fp, fn, tn))
```

```python
columns = ['threshold', 'tp', 'fp', 'fn', 'tn']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
df_scores[::10]
```

```python
df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)
df_scores
```

```python
plt.plot(df_scores.threshold, df_scores['tpr'], label='TPR')
plt.plot(df_scores.threshold, df_scores['fpr'], label='FPR')
plt.legend()
```

### Random model

```python
np.random.seed(1)
y_rand = np.random.uniform(0, 1, size=len(y_val))
```

```python
((y_rand >= 0.5) == y_val).mean()
```

```python
def tpr_fpr_dataframe(y_val, y_pred):
    scores = []
    thresholds = np.linspace(0, 1, 101)
    
    for t in thresholds:
        actual_positive = (y_val==1)
        actual_negative = (y_val==0)
    
        predict_positive = (y_pred >= t)
        predict_negative = (y_pred < t)
        
        tp = (predict_positive & actual_positive).sum()
        tn = (predict_negative & actual_negative).sum()
        fp = (predict_positive & actual_negative).sum()
        fn = (predict_negative & actual_positive).sum()
    
        scores.append((t, tp, fp, fn, tn))

    columns = ['threshold', 'tp', 'fp', 'fn', 'tn']
    df_scores = pd.DataFrame(scores, columns=columns)
    df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
    df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)
        
    return df_scores
```

```python
df_rand = tpr_fpr_dataframe(y_val, y_rand)
df_rand[::10]
```

```python
plt.plot(df_rand.threshold, df_rand['tpr'], label='TPR')
plt.plot(df_rand.threshold, df_rand['fpr'], label='FPR')
plt.legend()
```

### Cross validation

```python
### k-fold cross-validation 
# compute mean and std auc for the model

def train(df, y_train, C=1.0):
    dicts = df[categorical + numerical].to_dict(orient='records')

    dv = DictVectorizer(sparse=False)
    X_train = dv.fit_transform(dicts)

    model = lr(C=C, max_iter=1000)
    model.fit(X_train, y_train)

    return dv, model
    
```

```python
dv, model = train(df_train, y_train)
```

```python
def predict(df, dv, movel):
    dicts = df[categorical + numerical].to_dict(orient='records')

    X = dv.transform(dicts)
    y_pred = model.predict_proba(X)[:, 1]

    return y_pred

#c is inverse of regularisation, so the bigger c means less regularisaion
```

```python
import warnings
warnings.filterwarnings("ignore")
from tqdm.auto import tqdm

for c in [0.001, 0.01, 0.1, 1, 5 ,10]:
    scores = []
    for train_idx, val_idx in tqdm(kfold.split(df_full_train), total=10):
        df_train = df_full_train.iloc[train_idx]
        df_val = df_full_train.iloc[val_idx]
    
        y_train = df_train.churn.values
        y_val = df_val.churn.values
    
        dv, model = train(df_train, y_train, c)
        y_pred = predict(df_val, dv, model)
    
        auc = roc_auc_score(y_val, y_pred)
    
        scores.append(auc)
    print(f'{np.mean(scores):.3f}, {np.std(scores):.3f}')
```
